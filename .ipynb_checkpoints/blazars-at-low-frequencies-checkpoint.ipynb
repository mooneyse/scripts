# Blazars at low frequencies

* The bulk velocity is $ v $.
* The bulk Lorentz factor is $$ \gamma = \frac{1}{\sqrt{1 - \beta^{2}}} $$ where $ \beta = \frac{v}{c} $.
* The angle of the jet to our line of sight is $ \theta $.
* The specific intensity the $ e^{-} $ emit at is $ I_{\nu} \propto \nu^{\alpha} $ (a power law) between $ \nu_{1} $ and $ \nu_{2} $, and $ \frac{I_{\nu}}{\nu^{3}} $ is Lorentz invariant.

* The $ \nu $ of jet pointing at us is Doppler shifted (in frequency) by $ D_{+} = \gamma (1 + \beta \cos \theta) $. (The $ 1 + \beta \cos \theta $ is the non-relativistic Doppler shifting and the $ \gamma $ comes from time dilation.)
* The amplitude of the spectrum is boosted by $ D_{+}^{3} $, from the Lorentz invariant quantity above. The slope, $ \alpha $, does not change.
* The counterjet is deboosted by $ D_{-} = \gamma (1 - \beta \cos \theta) $.
* See this [Berkley tutorial](http://w.astro.berkeley.edu/~echiang/rad/ps8aans.pdf).
* A common way of finding the Doppler factor is using the equipartition of energy argument, the X-ray flux density, or (in [this](https://iopscience.iop.org/article/10.1086/307587) case), total flux density variation monitoring data at 22 GHz and 37 GHz.
* An important application of precise Doppler factors is calculating the Lorentz factors and viewing angles of AGN relativistic outflows.

```python
import numpy as np
import scipy.constants as sc
import matplotlib.pyplot as plt
import math
from IPython.display import Image

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

from matplotlib import pylab
pylab.rcParams['figure.figsize'] = (12, 8)
```

```python
def beta(v):
    ''' velocity '''
    return v / sc.c


def gamma(v):
    ''' lorentz factor '''
    return 1 / np.sqrt(1 - beta(v) ** 2)


def doppler(v, theta, jet='forward'):
    ''' doppler factor '''
    if jet == 'forward':
        return gamma(v) * (1 + beta(v) * np.cos(math.radians(theta)))
    elif jet == 'away':
        return gamma(v) * (1 - beta(v) * np.cos(math.radians(theta)))
    
    
def alpha(v, s, verbose=False):
    ''' spectral index '''
    a = np.log(s[0] / s[1]) / np.log(v[0] / v[1])
    if verbose:
        print('alpha =', np.round(a, 3))
    return a
    

def scale_flux(old_v, new_v, old_s, alpha):
    ''' scale the flux density using the spectral index relation '''
    return old_s * (new_v / old_v) ** alpha

    
def make_plot(x, y, x2='', y2='', x3='', y3='', xlabel='', ylabel='', Log=' '):
    ''' plot the data '''
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(x, y, color='w')
    if y2 != '':
        if x2 == '':
            x2 = x
        plt.plot(x2, y2, color='r')
    if y3 != '':
        if x3 == '':
            x3 = x
        plt.plot(x3, y3, color='b')
    if 'x' in Log:
        plt.xscale('log')
    if 'y' in Log:
        plt.yscale('log') 
    plt.show()
```

```python
fraction_of_c, lorentz_factors = [], []

for i in np.linspace(0, 1, 1000):
    if i != 1:  # prevents inf
        fraction_of_c.append(i)
        lorentz_factors.append(gamma(v=i * sc.c))

make_plot(x=fraction_of_c, y=lorentz_factors, xlabel='Fraction of $c$',
          ylabel='Lorentz factor')
```

```python
angle, doppler_forward, doppler_away = [], [], []

for i in np.linspace(0, 90, 1000):
    angle.append(i)
    doppler_forward.append(doppler(0.99 * sc.c, theta=i, jet='forward'))
    doppler_away.append(doppler(0.99 * sc.c, theta=i, jet='away'))

make_plot(x=angle, y=doppler_forward, y2=doppler_away, xlabel='Angle (degree)',
          ylabel='Doppler factor')
```

We have know the intrinsic values of $ I_{\nu} $ from a source at 10 MHz and 100 MHz from both jet lobes.

* $ I_{10+} = 10 $ Jy.
* $ I_{100+} = 5 $ Jy.

We will shift it for $ v = 0.95 c $ and $ \theta = 10 $ degrees.

```python
frequencies = np.array([10, 100])  # mhz
intensities = np.array([10, 5])  # jy

v = 0.95 * sc.c
theta = math.radians(10)

# forward jet
forward_frequencies = frequencies * doppler(v=v, theta=theta, jet='forward')
forward_intensities = intensities * doppler(v=v, theta=theta, jet='forward') ** 3

# away jet
away_frequencies = frequencies * doppler(v=v, theta=theta, jet='away')
away_intensities = intensities * doppler(v=v, theta=theta, jet='away') ** 3

make_plot(x=frequencies, y=intensities, x2=forward_frequencies, y2=forward_intensities,
          x3=away_frequencies, y3=away_intensities, xlabel=r'$\nu$ (MHz)',
          ylabel=r'$I_{\nu}$ (Jy)', Log='xy')

# spectral index does not change
old_alpha = alpha(v=frequencies, s=intensities, verbose=True)
forward_alpha = alpha(v=forward_frequencies, s=forward_intensities, verbose=True)
away_alpha = alpha(v=away_frequencies, s=away_intensities, verbose=True)
```

What is the difference in $ I_{\nu} $ between the jets at a given $ \nu $?

```python
new_intensities = []

# scale the counterjet to the forward jet frequencies
    
new_intensities.append(scale_flux(old_v=away_frequencies[0], new_v=forward_frequencies[0],
                                  old_s=away_intensities[0], alpha=old_alpha))
new_intensities.append(scale_flux(old_v=away_frequencies[1], new_v=forward_frequencies[1],
                                  old_s=away_intensities[1], alpha=old_alpha))

make_plot(x=forward_frequencies, y=new_intensities, x2=forward_frequencies,
          y2=forward_intensities, x3=away_frequencies, y3=away_intensities,
          xlabel=r'$\nu$ (MHz)', ylabel=r'$I_{\nu}$ (Jy)', Log='xy')

print('Intensity difference =', int(np.mean(forward_intensities) / np.mean(new_intensities)))
```

For M87, the ratio in brightness between the jets is 450 and $ \alpha = -0.5 $. i.e.
$$ \Big( \frac{D_{+}}{D_{-}} \Big)^{3 - \alpha} = 450 $$
where the $ \alpha $ is a correction factor arising because we scale. Then 
$$ \Big( \frac{\gamma (1 + \beta \cos \theta)}{\gamma (1 - \beta \cos \theta)} \Big)^{3 - \alpha} = 450 \\
\Big( \frac{1 + \beta \cos \theta}{1 - \beta \cos \theta} \Big)^{-3.5} = 450 $$

We constain the following:

* 45 degrees $ > \theta > $ 0 degrees by definition.
* $ \gamma > 1.4 $
* $ v > 0.7 c $


```python
def v_from_gamma(gamma):
    ''' get the constaint on the bulk velocity from the lorentz factor '''
    return np.sqrt(sc.c ** 2 - (sc.c / gamma) ** 2)/ sc.c

print('v >', str(np.round(v_from_gamma(1.4), 2)) + 'c')
```

## Burbidge's minimum energy argument

This argument gets a value for the magnetic field by minimising the total energy density.

This relies on a measurement of the volume emissivity, $ j_{\nu} $. The units are $ \mathrm{erg} \; \mathrm{s}^{-1} \; \mathrm{cm}^{-3} \; \mathrm{Hz}^{-1} \; \mathrm{sr}^{-1} $. It can be measured for optically thin media and we emasure it at the frequency at which the electrons radiate that are carrying the bulk of the kinetic energy density. The bulk energy density is carried by the lowest $ \gamma $ electrons because they are the most numerous. This holds so long as $ p < -2 $ where $$ \alpha = \frac{1 + p}{2}. $$

In the below image from [Carilli et al. (1991)](http://articles.adsabs.harvard.edu/pdf/1991ApJ...383..554C), we calculate $j_{\nu} $ at $ \nu = 10^{2.8} $ MHz as that is the lowest frequency for which the power law holds. $$ j_{\nu} \propto \frac{L_{\nu}}{V} $$ where $ V $ is the volume of the hotspot (we can assume a sphere) and $ L_{\nu} $ is the luminosity per hertz. To go from a sky scale to a radius, which is needed for the volume, we have to know the redshift. Likewise, the luminosity will also depend on the distance.

Doing this analysis gives $ B \approx 3 \times 10^{4} $ G (where 1 T = 10,000 G). Is it possible do find $ B $ with this line of reasoning for blazars? It requires the following:

* We need the volume emissivity, $ j_{\nu} $.
    * To find out what $ \nu $ to calculate $ j $ at, we need to ensure $ p < -2 $.
    * To get $ p $, we need the spectral index, $ \alpha $.
* $ j_{\nu} $ requires $ L_{\nu} $, the luminosity per hertz.
    * For $ L_{\nu} $, we need the flux density $ S_{\nu} $ (which has to be converted from $ \mathrm{Jy} \; \mathrm{beam}^{-1} $, requiring the beam size),
    * the luminosity distance $D_{L} $,
    * the redshift, $ z $,
    * and the spectral index, $ \alpha $ (already mentioned).
* $ j_{\nu} $ also requires the volume, $ V $.
    * $ V $ requires the radius, $ r $, if we assume a sphere.
    * $ r $ requires the distance, $ z $ (already mentioned).
    * $ r $ requires the size of the radio lobe on the radio map in arcseconds.
* Then $ j_{\nu} \approx n_{e} \cdot P_{\mathrm{single \; e^{-}}} \cdot \frac{1}{4 \pi} \cdot \frac{1}{\nu}$
